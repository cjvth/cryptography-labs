# Реализуйте LSB-matching (внедрение/извлечение).

## Внедрение

Отступим 1024 байта и дальше будем изменять каждый 97-ой байт

```python
import random
```

```python
with open('jab_jabich.bmp', 'rb') as f:
    original = bytearray(f.read())
```

```python
with open('message.txt', 'rb') as f:
    message1 = bytearray(f.read())
```

```python
def lsb_matching(byte: int, bit: int):
    if byte & 1 == bit:
        return byte
    if byte == 0:
        return byte + 1
    if byte == 255:
        return byte - 1
    if random.randint(0, 1) == 1:
        return byte + 1
    return byte - 1
```

```python
print(lsb_matching(0, 0))
print(lsb_matching(0, 1))
print(lsb_matching(255, 0))
print(lsb_matching(255, 1))
print()
print(lsb_matching(10, 1))
print(lsb_matching(10, 1))
print(lsb_matching(10, 1))
print(lsb_matching(10, 1))
print(lsb_matching(10, 1))
```

#### Само внедрение

```python
secret1 = original.copy()
assert(len(secret1) > len(message1) * 8 * 97 + 1024)
gen = iter(range(1024+97, len(secret1), 97))
for byte in message1 + b'\0':
    byte = int(byte)
    for i in range(8):
        bit = byte & 1
        byte >>= 1
        idx = next(gen)
        secret1[idx] = lsb_matching(secret1[idx], bit)
```

```python
with open('secret.bmp', 'wb') as f:
    f.write(secret1)
```

## Извлечение

```python
with open('secret.bmp', 'rb') as f:
    secret2 = bytearray(f.read())
```

```python
message2 = bytearray()
gen = iter(range(1024+97, len(secret2), 97))
while True:
    byte = 0
    for i in range(8):
        byte >>= 1
        idx = next(gen)
        bit = secret2[idx] & 1
        byte += bit << 7
    if byte == 0:
        break
    message2.append(byte)
```

```python
print(message2.decode('utf-8'))
```
