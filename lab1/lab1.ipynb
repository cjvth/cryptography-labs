## Функция Цезаря

```python
def caesar_do(text: str, key: int):
    def letter_do(letter: str, key: int):
        idx = ord(letter)
        if ord('a') <= idx <= ord('z'):
            idx = (idx - ord('a') + key) % 26 + ord('a')
        elif ord('A') <= idx <= ord('Z'):
            idx = (idx - ord('A') + key) % 26 + ord('A')
        return chr(idx)
    return ''.join(map(lambda letter: letter_do(letter, key), text))
```

```python
caesar_do('wxyzabcd', 3)
```

```python
caesar_do('wxyzabcd', -3)
```

## Словарь

```python
dictionary = set()
with open('words_alpha.txt') as file_dict:
    for word in file_dict.readlines():
        dictionary.add(word[:-1])
```

## Проверка словаря

```python
key = 17
with open('input.txt') as file_in:
    s = file_in.read()
s_decoded = caesar_do(s, -key)
s_decoded_words = re.findall(r'\w+', s_decoded)
s_decoded_words = [i.lower() for i in s_decoded_words]
for word in s_decoded_words[:5]:
    print(word, word in dictionary)
print()
for word in s_decoded_words:
    if word not in dictionary:
        print(word, word in dictionary)
```

## Перебор вариантов

```python
best_accuracy = 0
best_key = 0
with open('input.txt') as file_in:
    s = file_in.read()
total_words = len(re.findall(r'\w+', s))

for key in range(0, 26):
    s_decoded = caesar_do(s, -key)
    s_decoded_words = re.findall(r'\w+', s_decoded)
    s_decoded_words = [i.lower() for i in s_decoded_words]
    words_matched = 0
    for word in s_decoded_words:
        if word in dictionary:
            words_matched += 1
    accuracy = 100 * words_matched / total_words
    print(f'{key}\t{accuracy:.2f}')
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_key = key
```

```python
print(f'Best key: {best_key}')
decrypted = caesar_do(s, -17)
print('Decrypted with best key:')
print('\n==================\n')
print(decrypted)
with open('output.txt', 'w') as file_out:
    file_out.write(decrypted)
```
