## Подсчет символов

```python
def find_bytes_distribution(filename: str):
    a = [0] * 256
    with open(filename, 'rb') as f:
        data = f.read()
    total = len(data)
    for i in range(0, 256):
        a[i] = data.count(i) / total
    return a
```

```python
find_bytes_distribution('1_zeroes.txt')[ord('0') - 1: ord('0') + 2]
```

## Нахождение энтропии

```python
import math
def entropy_from_distribution(distribution):
    x = 0
    for p in distribution:
        if p > 0:
            x -= p * math.log2(p)
    return x
```

```python
entropy_from_distribution([0, 0, 1])
```

```python
entropy_from_distribution([0.5, 0.25, 0.25])
```

## Генерация файлов

```python
import random
random.seed(0)
with open('2_zeroes_ones.txt', 'wb') as f:
    f.write(bytes(random.choices(range(ord('0'), ord('1') + 1), k=1024)))
with open('3_letters.txt', 'wb') as f:
    f.write(bytes(random.choices(range(ord('a'), ord('z') + 1), k=1024)))
with open('4_printable.txt', 'wb') as f:
    f.write(bytes(random.choices(range(ord(' '), ord('~') + 1), k=1024)))
```

## Проверка энтропии

```python
for i in ['1_zeroes.txt', '2_zeroes_ones.txt', '3_letters.txt', '4_printable.txt', '../lab1/input_original.txt', 'lab2.ipynb']:
    print(i, f'{entropy_from_distribution(find_bytes_distribution(i)):.3f}')
```
