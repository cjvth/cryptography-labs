## Парадокс дней рождения

```python
import math
```

```python
def collision_probability(days, people):
    return 1 - math.factorial(days) / math.factorial(days - people) / (days ** people)
```

```python
print(f'Вероятность совпадения для 1 человека: {collision_probability(365, 1)*100:.2f}%')
print(f'Вероятность совпадения для 2 человек: {collision_probability(365, 2)*100:.2f}%')
print(f'Вероятность совпадения для 23 человек: {collision_probability(365, 23)*100:.2f}%')
print(f'Вероятность совпадения для 57 человек: {collision_probability(365, 57)*100:.2f}%')
```

### Если 256 дней

```python
print(f'Вероятность совпадения для 1 человека: {collision_probability(256, 1)*100:.2f}%')
print(f'Вероятность совпадения для 2 человек: {collision_probability(256, 2)*100:.2f}%')
print(f'Вероятность совпадения для 20 человек: {collision_probability(256, 20)*100:.2f}%')
print(f'Вероятность совпадения для 32 человек: {collision_probability(256, 32)*100:.2f}%')
```

### Проверка

```python
N = 100000
k = 20

collisions = 0
for i in range(N):
    datai = i.to_bytes(8, 'little')
    hashi = hashlib.sha256(datai).digest()
    if len(set(hashi[:k])) < k:
        collisions += 1
print(f'Экспериментальная вероятность для {k} человек: {collisions/N*100:.2f}%')
```
