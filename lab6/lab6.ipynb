# Реализуйте одностороннюю функцию быстрого возведения в степень

```python
def fast_pow(x, n):
    multiplications = 0
    result = 1
    current_pow = x
    while n > 0:
        if n & 1 == 1:
            result *= current_pow
            multiplications += 1
        n >>= 1
        current_pow *= current_pow
        multiplications += 1
    return result, multiplications
```

```python
for i in range(1, 10):
    print(f'2**{i}: {fast_pow(2, i)[0]}; {2 ** i}')
```

```python
for i in range(1, 10):
    print(f'3**{i}: {fast_pow(3, i)[0]}; {3 ** i}')
```

```python
for i in [15, 25, 35]:
    print(f'3**{i}: {fast_pow(3, i)[0]}; {3 ** i}')
```

## Сравнить количество выполненных умножений с весом Хэмминга

Кажется, что формула такая

multiplications(n) = floor(log2(n)) + bit_count(n) + 1

```python
import math
```

```python
for n in range(1, 35):
    multiplications = fast_pow(3, n)[1]
    log2 = int(math.log2(n))
    bits = n.bit_count()
    print(f'Ожидание: {log2}+{bits}+1={log2+bits+1}; эксперимент: {multiplications}')
```

```python
all_same = True
for n in range(1, 10000):
    multiplications = fast_pow(3, n)[1]
    log2 = int(math.log2(n))
    bits = n.bit_count()
    if log2+bits+1 != multiplications:
        all_same = False
        print('Отличается для n={n}')
if all_same:
    print('Всё совпало')
```
